```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = {
    'Hours': [2.5, 5.1, 3.2, 8.5, 3.5, 1.5, 9.2, 5.5, 8.3, 2.7, 7.7, 5.9, 4.5, 3.3, 1.1, 8.9, 2.5, 1.9, 6.1, 7.4, 2.7, 4.8, 3.8, 6.9, 7.8],
    'Scores': [21, 47, 27, 75, 30, 20, 88, 60, 81, 25, 85, 62, 41, 42, 17, 95, 30, 24, 67, 69, 30, 54, 35, 76, 86]
}
df = pd.DataFrame(data)
```

```python
X = df['Hours'].values.reshape(-1, 1)  # This is the Feature matrix
y = df['Scores'].values  # working on Target variable
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=52)
```

```python
X_train, X_test, y_train, y_test 
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.title('Actual vs. Predicted Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.show()
```

```python
hours = float(input("Enter the number of study hours: "))
percentage = model.predict([[hours]])
print("Predicted percentage based on {} hours of study: {:.2f}".format(hours, percentage[0]))
```
